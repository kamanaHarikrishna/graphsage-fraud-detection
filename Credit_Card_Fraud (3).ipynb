```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb

# Drop target column and split
X = df.drop(columns='Class')
y = df['Class']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from sklearn.model_selection import GridSearchCV


scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

# Step 3: Define parameter grid for tuning
param_grid = {
    'max_depth': [4, 6],
    'learning_rate': [0.05, 0.1],
    'n_estimators': [100, 150],
    'scale_pos_weight': [scale_pos_weight]  # Handling class imbalance
}

# Step 4: Initialize model
xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Step 5: Grid search
grid = GridSearchCV(estimator=xgb_clf, param_grid=param_grid, cv=3, scoring='f1', verbose=1, n_jobs=-1)
grid.fit(X_train, y_train)

# Step 6: Best model
best_model = grid.best_estimator_

# Step 7: Evaluation
y_pred = best_model.predict(X_test)
print("Best Parameters:", grid.best_params_)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
#print("ROC-AUC Score:", roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1]))

# Step 8: Confusion Matrix Plot
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("XGBoost Confusion Matrix (Tuned)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Predict probabilities
y_probs = best_model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='XGBoost (AUC = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Random guess line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
import torch
from torch_geometric.data import Data
import networkx as nx

# Use a subset for memory efficiency
subset_df = df.head(10000).copy()

# Create graph
G = nx.Graph()
for idx, row in subset_df.iterrows():
    G.add_node(idx, features=row.drop('Class').values, label=int(row['Class']))

# Add edges based on similarity (you can customize thresholds)
for i in range(len(subset_df)):
    for j in range(i+1, min(i+100, len(subset_df))):
        amt_diff = abs(subset_df.iloc[i]['Amount'] - subset_df.iloc[j]['Amount'])
        time_diff = abs(subset_df.iloc[i]['Time'] - subset_df.iloc[j]['Time'])
        if amt_diff <= 0.01 or time_diff <= 5:
            G.add_edge(i, j)

# Convert to PyTorch Geometric format
features = torch.tensor([G.nodes[n]['features'] for n in G.nodes], dtype=torch.float)
labels = torch.tensor([G.nodes[n]['label'] for n in G.nodes], dtype=torch.long)
edge_index = torch.tensor(list(G.edges)).t().contiguous()

data = Data(x=features, edge_index=edge_index, y=labels)
```

```python
from torch_geometric.nn import SAGEConv  # You can swap with GATConv for GAT
import torch.nn.functional as F
import torch.nn as nn

class GraphSAGE(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, dropout=0.3):
        super(GraphSAGE, self).__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, out_channels)
        self.dropout = dropout

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Compute class weights for NLLLoss
weights = compute_class_weight(class_weight='balanced', classes=np.unique(labels.numpy()), y=labels.numpy())
class_weights = torch.tensor(weights, dtype=torch.float)

# Define loss with weights
loss_fn = nn.NLLLoss(weight=class_weights)
```

```python
model = GraphSAGE(
    in_channels=data.num_node_features,
    hidden_channels=32,
    out_channels=2
)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Training loop
model.train()
epochs = 50

for epoch in range(epochs):
    optimizer.zero_grad()
    out = model(data)
    loss = loss_fn(out, data.y)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}/{epochs} | Loss: {loss.item():.4f}")
```

```python
from sklearn.metrics import classification_report, accuracy_score
import torch

# Set model to evaluation mode
model.eval()

# Get predictions
_, predictions = model(data).max(dim=1)

# Convert predictions and labels to numpy arrays
y_true = data.y.cpu().numpy()
y_pred = predictions.cpu().numpy()

# Generate classification report
report = classification_report(y_true, y_pred, digits=4)
print("\n📊 Improved GraphSAGE GNN Report (with class weighting):")
print(report)

# Print accuracy explicitly
acc = accuracy_score(y_true, y_pred)
print(f"\n✅ Accuracy: {acc:.4f}")

# Optional: Get report as a dictionary (for plots or saving to file)
report_dict = classification_report(y_true, y_pred, output_dict=True)

# Example: Extract key metrics
fraud_precision = report_dict['1']['precision']
fraud_recall = report_dict['1']['recall']
fraud_f1 = report_dict['1']['f1-score']
macro_avg_f1 = report_dict['macro avg']['f1-score']
weighted_avg_f1 = report_dict['weighted avg']['f1-score']

# Print extracted metrics (optional)
print("\n🔍 Key Metrics:")
print(f"Fraud Precision: {fraud_precision:.4f}")
print(f"Fraud Recall:    {fraud_recall:.4f}")
print(f"Fraud F1-Score:  {fraud_f1:.4f}")
print(f"Macro Avg F1:    {macro_avg_f1:.4f}")
print(f"Weighted Avg F1: {weighted_avg_f1:.4f}")
```
